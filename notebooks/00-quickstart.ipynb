# MOTIF Quick StartGet up and running with MOTIF in 5 minutes!

```python
import sys
sys.path.insert(0, '..')

import numpy as np
from src.synthetic_data import generate_cohort
from src.motif_pipeline import calibrate_patient, generate_motif_proxies, extract_features
```

## Generate a 20-patient cohortCreate a small stratified cohort across the three inflammation outcomes.

```python
print('Generating 20-patient cohort...')cohort = generate_cohort(N_resolution=7, N_chronic=6, N_death=7, seed=42, verbose=False)print(f'Generated {len(cohort)} patients')
```

## Calibrate the cohort with MOTIFFit ODE parameters to each patient's observed data.

```python
print('Running MOTIF calibration pipeline...')results = []for i, patient in enumerate(cohort):    if i % 5 == 0:        print(f'  Calibrating patient {i+1}/{len(cohort)}...')    fitted_params, fitted_P0, fit_quality = calibrate_patient(patient, n_restarts=1, verbose=False)    results.append({        'patient_id': patient['id'],        'outcome': patient['outcome'],        'fit_quality': fit_quality,        'fitted_params': fitted_params,        'fitted_P0': fitted_P0    })print('Pipeline complete')n_calibrated = len(results)print(f'Calibrated {n_calibrated} patients')
```

## View calibration resultsDisplay R-squared values to assess how well the model fits the data.

```python
import pandas as pdresults_df = pd.DataFrame([    {'Patient': r['patient_id'], 'Outcome': r['outcome'], 'R2': r['fit_quality']}    for r in results])print('Calibration Quality (R-squared):')print(results_df.to_string(index=False))print(f"\nMean R2: {np.mean([r['fit_quality'] for r in results]):.4f}")print(f"Median R2: {np.median([r['fit_quality'] for r in results]):.4f}")print(f"Min R2: {np.min([r['fit_quality'] for r in results]):.4f}")print(f"Max R2: {np.max([r['fit_quality'] for r in results]):.4f}")
```
